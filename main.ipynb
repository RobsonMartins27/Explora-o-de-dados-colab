# Correções

```python
# customers - age - valor

media_idade = df_customers['age'].mean()
df_customers['age'] = df_customers['age'].fillna(media_idade)
display(df_customers)
```

```python
# rentals - daily_rate coom valor menor que 0
df_rentals.loc[df_rentals['daily_rate']<=0, 'daily_rate'] = pd.NA
display(df_rentals)
```

```python
# vehicles - co2_g_per_km - dado ausente
media_co2 = df_vehicles['co2_g_per_km'].mean()
df_vehicles['co2_g_per_km'] = df_vehicles['co2_g_per_km'].fillna(media_co2)
display(df_vehicles)
```

```python
# rentals - daily rate - valor como string
df_rentals['daily_rate'] = pd.to_numeric(df_rentals['daily_rate'], errors='coerce')
display(df_rentals)
```

```python
# rentals - data invalida
df_rentals['start_date'] = pd.to_datetime(df_rentals['start_date'], errors='coerce')
display(df_rentals)
```

```python
# rentals - data com formatação diferente
df_rentals['end_date'] = pd.to_datetime(df_rentals['end_date'], errors='coerce')
display(df_rentals)
```

# Engenharia de dados


- Calcule a duração de cada locação (em dias).

- Calcule a receita total por locação

- Faça o merge com a base de veículos

- Calcule a emissão de CO2 po locação

- Extraia ano e mês da data de início de locação

```python
# criar backup de dados
df_carbon_clean = df_carbon
df_rentals_clean = df_rentals
df_customers_clean = df_customers
df_vehicles_clean = df_vehicles
```

```python
# Calcule a duração de cada locação (em dias).
df_rentals_clean['dias_locação'] = (df_rentals_clean['end_date'] - df_rentals_clean['start_date']).dt.days
display(df_rentals_clean)

```

```python
# Calcule a receita total por locação
df_rentals_clean['receita_total'] = (df_rentals_clean['dias_locação'] * df_rentals_clean['daily_rate'])
display(df_rentals_clean)
```

```python
# Faça o merge com a base de veículos

df_consolidado = df_rentals_clean.merge(df_vehicles_clean, on='vehicle_id')
df_consolidado = df_consolidado.merge(df_customers_clean, on='customer_id')
```

```python
# Calcule a emissão de CO2 po locação
df_consolidado['Co2_emitido'] = df_consolidado['co2_g_per_km'] * df_consolidado['distance_km']
```

```python
# Extraia ano e mês da data de início de locação
df_consolidado['mes'] = df_consolidado['start_date'].dt.month
df_consolidado['ano'] = df_consolidado['start_date'].dt.year
df_consolidado['dia_da_semana'] = df_consolidado['start_date'].dt.weekday

display(df_consolidado)

```
